```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/korla/Desktop/krish codes/simple-Linear-Regression-master/Salary_Data.csv")
df.head()
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,1].values
```

```python
# spliting the data into Training and test in 70 - 30 format
```

```python
from IPython.display import Image
Image("C://Users//korla//Downloads//Televzr Downloads//minist 3.PNG",width=800,height=800)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# X_train and y_train values
X_train , y_train
```

```python
# X_test and y_test values
X_test , y_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
# In equation y = mx + c
```

```python
# intercept value

reg.intercept_
```

```python
# coefficient value

reg.coef_
```

```python
from IPython.display import Image
Image("C://Users//korla//Desktop//img//simple Linear regression 3.JPG",width=600,height=600)
```

```python
# predicted values

X_train_pred=reg.predict(X_train)
```

```python
X_train_pred
```

```python
#analysing the actual and predicted values

a=pd.DataFrame({'actual values':y_train,'precict values':X_train_pred})
a
```

```python
%matplotlib inline
plt.xlabel("Years of Experience")
plt.ylabel("salary")
plt.title("Experience vs Salary")
plt.scatter(X_train,y_train,color="red",marker="+")
plt.plot(X_train,X_train_pred,color='blue',marker='o')
```

```python
%matplotlib inline
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("checking")
plt.scatter(X_test,y_test,color="red",marker="+")
plt.plot(X_train,X_train_pred,color='blue',marker='o')
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,X_train_pred) * 100
```

```python
y_test_pred = reg.predict(X_test)
y_test_pred
```

```python
r2_score(y_test,y_test_pred)
```
